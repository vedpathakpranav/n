Co-reference resolution model: Implement a co-reference resolution model to identify and link
pronouns or noun phrases to the entities they refer to within a given text.

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from fastcoref import FCoref
import re

model = FCoref(device='cpu')
text = """
Ravi met Suresh in the office. He greeted him warmly.
They discussed the project.
"""

preds = model.predict(texts=[text])
result = preds[0]

clusters = result.get_clusters(as_strings=True)

print("----Coreference CLusters------")
for i,clusters in enumerate(clusters):
    print(f"Cluster {i+1}: {clusters}")

print("\n----Resolved Text---------")
resolved_text= text

for cluster in clusters:
    main_entity = clusters[0]

    for mention in clusters[1:]:
        pattern = r'\b{}\b'.format(re.escape(mention))
        resolved_text = re.sub(pattern,main_entity,resolved_text)
        
print(resolved_text)
```
