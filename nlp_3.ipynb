Viterbi Decoding Algorithm for POS: Given a sequence of words, implement the Viterbi algorithm
to find the most likely sequence of POS tags by estimating emission and transition probabilities from a
tagged corpus.

## Step 1: Import Library

```python
import numpy as np
```

## Step 2: Define Input

```python
sentence = ["time", "flies"]
tags = ["N","V"]
```

## Step 3: Initial Probabilities

```python
pi = {"N":0.6, "V":0.4}
```

## Step 4: Transition Probabilities

```python
transition = {
    "N": {"N":0.3, "V":0.7},
    "V": {"N":0.8, "V":0.2}
}
```

## Step 5: Emission Probabilities

```python
emission = {
    "time": {"N":0.5, "V":0.1},
    "flies": {"N":0.2, "V":0.6}
}
```

## Step 6: Create Tables

```python
V = [{} for _ in range(len(sentence))]
path = {}
```

## Step 7: Initialization (First Word)

```python
for tag in tags:
    V[0][tag] = pi[tag] * emission[sentence[0]][tag]
    path[tag] = [tag]
```

## Step 8: Recursion

```python
for t in range(1, len(sentence)):
    new_path = {}
    
    for curr in tags:
        prob, prev = max(
            (V[t-1][p] * transition[p][curr] * emission[sentence[t]][curr], p)
            for p in tags
        )
        
        V[t][curr] = prob
        new_path[curr] = path[prev] + [curr]
    
    path = new_path
```

## Step 9: Final Result

```python
final_tag = max(V[-1], key=V[-1].get)
sequence = path[final_tag]
```

## Step 10: Print Output

```python
print("Best Sequence: ", sequence)
print("Final Probability: ", V[-1][final_tag])
```
