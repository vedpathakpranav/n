## Step 2: Take Sample Text

```python
text = "Natural Language Processing (NLP) is amazing! It helps computers understand human language, isn't it interesting?"
```

```python
print(text)
```

## Step 3: Convert Uppercase to Lowercase

```python
lower_text = text.lower()
```

```python
print(lower_text)
```

## Step 4: Remove Punctuation

```python
import string
```

```python
no_punct_text = ""
```

```python
for char in lower_text:
    if char not in string.punctuation:
        no_punct_text = no_punct_text + char

print(no_punct_text)
```

## Step 5: Sentence Tokenization

```python
from nltk.tokenize import sent_tokenize
```

```python
sentences = sent_tokenize(no_punct_text)
print(sentences)
```

## Step 6: Word Tokenization

```python
from nltk.tokenize import word_tokenize
```

```python
words = word_tokenize(no_punct_text)
print(words)
```

## Step 7: Stopword Removal

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
clean_words = []
```

```python
for word in words:
    if word not in stop_words:
        clean_words.append(word)

print(clean_words)
```

## Step 8: Stemming (Porter Stemmer)

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
stemmed_words = []
```

```python
for word in clean_words:
    stemmed_words.append(ps.stem(word))

print(stemmed_words)
```

## Step 9: Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatized_words = []
```

```python
for word in clean_words:
    lemmatized_words.append(lemmatizer.lemmatize(word))

print(lemmatized_words)
```
