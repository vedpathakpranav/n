## Load data

```python
import json
```

```python
file_path = "C:/Users/PRANAV/Downloads/cellphone/Cell_Phones_and_Accessories_5.json"
```

```python
reviews = []
```

```python
with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        reviews.append(json.loads(line))
```

```python
print("Total reviews:", len(reviews))
```

## Preprocessing

```python
import re
```

```python
from nltk.tokenize import word_tokenize
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words("english"))
```

```python
sentences = []
```

```python
for r in reviews:
    text = r.get("reviewText", "").lower()
    
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    words = word_tokenize(text)
    
    words = [w for w in words if w not in stop_words]
    
    if len(words) > 3:
        sentences.append(words)

print("Sample sentence:", sentences[0])
```

## Word2Vec

```python
from gensim.models import Word2Vec
```

```python
w2v = Word2Vec(sentences, vector_size=100, window=5, min_count=5)
```

```python
print("\nWord2Vec similar to 'battery':")
w2v.wv.most_similar("battery")
```

```python
print("\nWord2Vec similarity (movie, netflix):")
print(w2v.wv.similarity("movie", "netflix"))
```

## FastText

```python
from gensim.models import FastText
```

```python
ft = FastText(sentences, vector_size=100, window=5, min_count=5)
```

```python
print("\nFastText similar to misspelled 'battrey':")
ft.wv.most_similar("battrey")
```

```python
print("\nFastText similarity (movie, netflix):")
print(ft.wv.similarity("movie", "netflix"))
```

## Save Models

```python
w2v.save("word2vec.model")
ft.save("fasttext.model")
```
