Spell checking and correction: Apply minimum edit distance between two strings for spelling
correction.

## Step 1: Input

```python
str1 = input("Enter First String: ")
str2 = input("Enter Second String: ")
```

```python
rows = len(str1)
cols = len(str2)
```

## Step 2: Create Matrix

```python
T = [[0 for j in range(cols+1)] for i in range(rows+1)]
```

## Step 3: Fill first row and column

```python
for i in range(rows + 1):
    T[i][0] = i

for j in range(cols + 1):
    T[0][j] = j
```

## Step 4: Fill matrix

```python
for i in range(1, rows+1):
    for j in range(1, cols+1):

        if str1[i-1] == str2[j-1]:
            T[i][j] = T[i-1][j-1]
        else:
            T[i][j] = min(
                T[i-1][j] + 1,      # delete
                T[i][j-1] + 1,      # insert
                T[i-1][j-1] + 1     # replace
            )
```

## Step 5: Count operations

```python
i = rows
j = cols

insert = delete = replace = 0

while i > 0 or j > 0:

    if i > 0 and j > 0 and str1[i-1] == str2[j-1]:
        i -= 1
        j -= 1

    elif i > 0 and j > 0 and T[i][j] == T[i-1][j-1] + 1:
        replace += 1
        i -= 1
        j -= 1

    elif i > 0 and T[i][j] == T[i-1][j] + 1:
        delete += 1
        i -= 1

    else:
        insert += 1
        j -= 1
```

## Step 6: Print Matrix

```python
print("\nEdit Distance Matrix:")
for row in T:
    print(row)
```

```python
print("\nMinimum Edit Distance:", T[rows][cols])

print("\nOperation Counts:")
print("Insert operations:", insert)
print("Remove operations:", delete)
print("Replace operations:", replace)
```
